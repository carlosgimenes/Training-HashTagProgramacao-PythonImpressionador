#### 12. Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês.

```python
salario_hora = float(input("Qual seu salário por hora: "))
horas_mes = float(input("Quantas horas trabalhou no mês: "))
```

#####  Calcule o salário bruto (horas * salario por hora)

```python
salario_bruto = horas_mes * salario_hora

print("O valor do salário bruto é R${}".format(salario_bruto))
```

##### Calcule o desconto do IR (11% do salário bruto)

```python
irpf = 0.11
desconto_irpf = salario_bruto * irpf

print("O Imposto de Renda para este salário é R${}".format(desconto_irpf))
```

##### Calcule o desconto do INSS (8% do salário bruto)

```python
inss = 0.08
desconto_inss = salario_bruto * inss

print("O INSS para este salário é R${}".format(desconto_inss))
```

##### Calcule o desconto do sindicato (5% do salário bruto)

```python
sindicato = 0.05
desconto_sindicato = salario_bruto * sindicato

print("O valor da Contribuição Sindical é R${}".format(desconto_sindicato))
```

##### Calcule o salário líquido (salário bruto - descontos)

```python
descontos = desconto_irpf + desconto_inss + desconto_sindicato
salario_liquido = salario_bruto - descontos

print("O valor do salário após os descontos é de R${}".format(salario_liquido))
```
